```python
def get_representative_waveform(WFs):
    WFs_rep=np.zeros((len(WFs.keys()),82))
    i=0
    for k in WFs.keys():
        unit_WFs=np.array(WFs[k])
        WF_max_amp=np.min(unit_WFs)
        row_ind=np.where(unit_WFs==WF_max_amp)[0]
        WFs_rep[i,:]=unit_WFs[row_ind]
        i+=1
    return WFs_rep

unnormWFs=get_representative_waveform(WFs)
```
